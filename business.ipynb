Proyecto 3: Optimización de Campañas de Telemarketing Bancario

Descripción General
Dataset real de un banco portugués que ejecutó campañas telefónicas de venta directa entre mayo 2008 y noviembre 2010. El período coincide con la crisis financiera global, lo que convierte a los indicadores macroeconómicos en variables críticas del análisis. El dataset fue publicado por Moro et al. en el UCI ML Repository y es ampliamente utilizado en benchmarking de modelos de clasificación en contextos financieros.

Fuente: UCI ML Repository
Registros: 41,188
Variables: 21
Target: y (yes/no) — tasa de suscripción base ~11% (dataset desbalanceado)
Tipo: Clasificación binaria


Contexto de Negocio
El banco busca maximizar la efectividad de sus campañas telefónicas para la suscripción a depósitos a plazo fijo. El objetivo no es únicamente maximizar la tasa de conversión, sino reducir el número total de llamadas necesarias para alcanzar las mismas o mejores suscripciones, mejorando directamente el ROI de la campaña y la eficiencia del call center.

Variables Clave
Variables del cliente (demográficas):

age: edad del cliente
job: tipo de empleo
marital: estado civil
education: nivel educativo
default: si tiene crédito en mora
housing: si tiene préstamo hipotecario
loan: si tiene préstamo personal

Variables de contacto e historial de campaña:

contact: tipo de comunicación (celular/teléfono)
duration: duración de la última llamada en segundos (*)
campaign: número de contactos en la campaña actual
pdays: días transcurridos desde el último contacto en campaña anterior
previous: número de contactos antes de esta campaña
poutcome: resultado de la campaña anterior

Variables macroeconómicas:

euribor3m: tasa Euribor a 3 meses
cons.conf.idx: índice de confianza del consumidor
emp.var.rate: tasa de variación del empleo
cons.price.idx: índice de precios al consumidor
nr.employed: número de empleados (indicador trimestral)

Variable objetivo:

y: ¿el cliente suscribió un depósito a plazo fijo? (yes/no)

(*) Nota importante: duration es el predictor más fuerte estadísticamente, pero representa una fuga de datos para modelos predictivos operativos, ya que solo se conoce al finalizar la llamada. Debe tratarse con cuidado en el modelado.

Preguntas de Negocio

¿Qué perfil demográfico (edad, empleo, educación) presenta la mayor tasa de conversión?
¿Cuántas llamadas son demasiadas? ¿Existe un punto de retorno decreciente según el número de contactos realizados?
¿Qué tan decisiva es la duración de la llamada como predictor? ¿Puede usarse en un modelo operativo real?
¿Los clientes contactados previamente con éxito (poutcome = success) tienen una tasa de conversión significativamente mayor?
¿Qué rol juega el contexto macroeconómico (Euribor, tasa de empleo) en la decisión del cliente?
¿Qué porcentaje de la base de clientes puede descartarse sin perder más del 5% de las conversiones potenciales?


KPIs de Éxito del Modelo

Precision en clase positiva >= 70%: de los clientes marcados como probables suscriptores, al menos 7 de 10 deben convertir. Reduce llamadas innecesarias.
Recall en clase positiva >= 60%: el modelo debe capturar al menos el 60% de los clientes que sí suscribirían, evitando oportunidades perdidas.
AUC-ROC >= 0.80: métrica global de discriminación independiente del umbral de clasificación.
Reducción de volumen de llamadas >= 30%: si el modelo filtra el 30% o más de llamadas improductivas, el ROI de campaña mejora directamente.
F1-Score como criterio de comparación entre modelos: dado el desbalance de clases (~11% positivo), el F1 es más representativo que el accuracy simple.
