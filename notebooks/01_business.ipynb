# PROYECTO TELEMARKETING BANCARIO
## PB-01: Entendimiento del Negocio (Business Understanding)
### 1. Contexto Comercial
Las campañas de telemarketing directo representan uno de los canales más efectivos para la captación de pasivos (suscripción de depósitos a plazo fijo) en el sector bancario. Sin embargo, la ejecución masiva de llamadas (estrategia de fuerza bruta) presenta dos grandes problemas: genera altos costos operativos continuos en el Call Center y deteriora la experiencia del usuario al interrumpir a clientes sin propensión de compra.

### 2. Definición del Problema Analítico
Actualmente, el equipo de ventas invierte una cantidad desproporcionada de horas-hombre contactando a una base de clientes no segmentada. El reto consiste en transformar este enfoque masivo en una estrategia dirigida, utilizando el historial de campañas anteriores (41,188 registros) para predecir el comportamiento futuro de los clientes.

### 3. Objetivos del Proyecto
Objetivo de Negocio: Optimizar la eficiencia de las campañas de telemarketing, priorizando los esfuerzos de contacto hacia los perfiles con mayor probabilidad de conversión.

Objetivo de Data Science: Desarrollar y validar un modelo de clasificación binaria capaz de predecir si un cliente suscribirá (Sí/No) un depósito a plazo fijo tras ser contactado.

### 4. KPIs e Impacto Financiero Esperado
La implementación de este modelo busca impactar directamente en tres métricas clave del banco:

*   Reducción del Costo Operativo (OPEX): Disminuir el volumen de llamadas infructuosas en un 15%. Esto se logra al filtrar la base de contactos y priorizar a aquellos con mayor probabilidad de conversión, optimizando el tiempo de los operadores.

*   Incremento del Ratio de Conversión: Aumentar el número de depósitos a plazo fijo colocados por cada 100 llamadas realizadas, maximizando el Retorno de Inversión (ROI) de la campaña de telemarketing.

*   Protección del NPS (Net Promoter Score): Reducir la "fatiga de contacto" evitando llamar a clientes cuyo perfil indica un claro rechazo al producto.

### 5. Traducción de Negocio a Machine Learning
Para asegurar que el modelo sirva a los intereses del banco, definimos las siguientes directrices técnicas:

Variable Objetivo (Target): La columna y del dataset (1 = Suscribe, 0 = No suscribe).

Métrica de Éxito Principal: Nos enfocaremos en optimizar el Recall (Sensibilidad) de la clase positiva y el F1-Score.

Justificación de la Métrica: En este escenario bancario, las clases están fuertemente desbalanceadas (la mayoría de clientes dice "No"). Utilizar el Accuracy (Exactitud) sería engañoso. Para el banco, el costo de oportunidad de dejar pasar a un cliente que sí quería invertir (Falso Negativo) es mucho mayor que el costo de hacer una llamada de unos minutos a alguien que terminará rechazando la oferta (Falso Positivo). Por tanto, priorizamos "capturar" la mayor cantidad posible de clientes interesados.

## PB-02: Criterios de Éxito del Proyecto

1.   Criterio de Negocio (Económico): Lograr una reducción del 10% en el costo operativo de las campañas de telemarketing. Esto se medirá comparando la tasa de éxito actual (columna y) contra la proyectada al priorizar registros con alta probabilidad de conversión.

2.   Criterio Estadístico (Machine Learning): El modelo debe alcanzar un Recall superior al 70% para la clase positiva (y = 'yes'). Esto garantiza que identifiquemos a la gran mayoría de clientes interesados sin desperdiciar recursos en llamadas fallidas.

3.   Relación con la Data: El modelo utilizará variables demográficas (como age, job, education) y de comportamiento previo (campaign, previous, poutcome) para predecir si el cliente aceptará el producto bancario (variable objetivo y).
