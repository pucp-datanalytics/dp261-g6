Cambiamos 999 por -1 para que el modelo entienda que son clientes nuevos y no se confunda con un número gigante.
Esto evita que el algoritmo distorsione los promedios con datos de personas que nunca han sido contactadas.

```python
df['pdays'] = df['pdays'].replace(999, -1)
```

Borramos esta columna porque no sabremos cuánto dura una llamada hasta que ya terminó; incluirla sería "hacer trampa".
Al quitarla, evitamos el Data Leakage y aseguramos que el modelo haga predicciones realistas.

```python
df = df.drop(columns=['duration'])
```

Simplificamos el historial a un "sí" o "no" para saber rápidamente si el cliente ya conoce la marca.
Esto ayuda al modelo a capturar el efecto de la familiaridad del cliente de forma más directa.

```python
df['contacted_before'] = (df['previous'] > 0).astype(int)
```

Medimos qué tan "pesados" somos con el cliente y ponemos un tope de 10 para que los casos exagerados no rompan el cálculo.
De esta forma controlamos los valores extremos (outliers) y mantenemos estable el aprendizaje del algoritmo.

```python
df['campaign_intensity'] = df['campaign'] / (df['previous'] + 1)
df['campaign_intensity'] = df['campaign_intensity'].clip(0, 10)
```

Juntamos las deudas en una sola variable para ver, de un vistazo, si el cliente ya tiene el bolsillo comprometido.
Simplifica la información financiera para que el modelo identifique más fácil el nivel de endeudamiento total.

```python
df['has_loan_or_housing'] = (
    (df['loan_yes'] == 1) | (df['housing_yes'] == 1)
).astype(int)
```
