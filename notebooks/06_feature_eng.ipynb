```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/03-bank_marketing.csv', sep=';')
```

```python
df.describe()
```

```python
df['pdays'] = df['pdays'].replace(999, -1)
```

```python
cols = ['job','marital','education','default','housing','loan']
for col in cols:
    df[col] = df[col].replace('unknown', 'other')
```

```python
df = df.drop(columns=['duration']) ##se borra porque impacta directamente en la variable objetivo, una llamada larga se puede asociar a que si acepta el deposito, pero esto no es posible porque contiene informacion posterior a lo que se requiere predecir
```

```python
df['y'] = df['y'].map({'yes':1,'no':0})
```

```python
df['contacted_before'] = (df['previous'] > 0).astype(int)
```

```python
df['campaign_intensity'] = df['campaign'] / (df['previous'] + 1)
df['campaign_intensity'] = df['campaign_intensity'].clip(0, 10)
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[18,30,45,60,100],
    labels=['young','adult','mid','senior']
)
```

```python
df['has_loan_or_housing'] = (
    (df['loan'] == 'yes') | (df['housing'] == 'yes')
).astype(int)
```

```python
df = pd.get_dummies(df, drop_first=True).astype(int)
```
