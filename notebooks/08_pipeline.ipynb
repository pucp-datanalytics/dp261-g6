```python
import pandas as pd
import joblib
import os
import sys

# Configurar rutas para importar desde src
sys.path.append(os.path.abspath('../'))
from src.preprocessing import get_preprocessor, custom_feature_engineering

# --- PASO 1: Cargar tu trabajo previo (Notebook 07) ---
df = pd.read_csv('../data/processed/train_balanced.csv')

# --- PASO 2: Aplicar ingeniería de variables (Notebook 06) ---
df = custom_feature_engineering(df)

# --- PASO 3: Definir X e y ---
X_train = df.drop(columns=['y'])
y_train = df['y']

# --- PASO 4: Configurar Columnas
num_cols = X_train.select_dtypes(include=['number']).columns.tolist()
cat_cols = X_train.select_dtypes(include=['object', 'category']).columns.tolist()

# --- PASO 5: Crear e integrar el Pipeline
preproc = get_preprocessor(num_cols, cat_cols)

# Validamos que se entrena (fit) y aplica (transform) sin errores
X_train_transformed = preproc.fit_transform(X_train)
print(f"Éxito: El pipeline generó una matriz de forma: {X_train_transformed.shape}")

# --- PASO 6: Persistir el Pipeline
os.makedirs('../models', exist_ok=True)
joblib.dump(preproc, '../models/preprocessor.pkl')
print("Pipeline guardado en '../models/preprocessor.pkl'")

# --- PASO 7: Guardar para DVC
# Convertimos a DataFrame para que el archivo final sea legible
df_final = pd.DataFrame(X_train_transformed, columns=preproc.get_feature_names_out())
df_final['target'] = y_train.values

os.makedirs('../data/final', exist_ok=True)
df_final.to_csv('../data/final/train_final_transformed.csv', index=False)
print("Dataset procesado guardado en '../data/final/train_final_transformed.csv'")
```
