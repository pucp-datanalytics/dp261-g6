```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Configuración de visualización
sns.set_theme(style="whitegrid")
%matplotlib inline

# Cargar el dataset (ajustando la ruta desde la carpeta notebooks)
df = pd.read_csv('../data/raw/03-bank_marketing.csv', sep=';')

# Dimensiones del dataset
print(f"Registros: {df.shape[0]}")
print(f"Variables: {df.shape[1]}")
```

```python
df.info()
```

```python
# Tipos y Nulos
info_df = pd.DataFrame({
    'Tipo': df.dtypes,
    'Nulos': df.isnull().sum(),
    '% Nulos': (df.isnull().sum() / len(df)) * 100
})
print(info_df)

# Duplicados
print(f"Registros duplicados: {df.duplicated().sum()}")

# Outliers (Visualización rápida para variables numéricas clave)
num_cols = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(15, 5))
df[num_cols].boxplot()
plt.xticks(rotation=45)
plt.title("Detección de Outliers")
plt.show()
```

```python
# Estadísticas descriptivas generales
print(df.describe())

# Balance de la variable target ('y')
print("\nBalance de Clases:")
print(df['y'].value_counts())
print(df['y'].value_counts(normalize=True) * 100)

# Gráfico de balance
sns.countplot(x='y', data=df, palette='viridis')
plt.title("Distribución del Target (Balance de Clases)")
plt.show()
```

```python
# Histogramas de variables numéricas
df.hist(figsize=(15, 10), bins=30, color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()

# Heatmap de Correlación
plt.figure(figsize=(12, 8))
# Convertimos el target a numérico (yes=1, no=0) para ver correlación
df_corr = df.copy()
df_corr['y'] = df_corr['y'].map({'yes': 1, 'no': 0})
sns.heatmap(df_corr.corr(numeric_only=True), annot=True, cmap='RdYlGn', fmt=".2f")
plt.title("Mapa de Calor de Correlaciones")
plt.show()
```

```python
# Relación de variables categóricas con el Target (Ejemplo: Educación)
plt.figure(figsize=(12, 6))
sns.countplot(x='education', hue='y', data=df)
plt.xticks(rotation=45)
plt.title("Suscripción según Nivel Educativo")
plt.show()

# Boxplot: Edad vs Target
plt.figure(figsize=(8, 6))
sns.boxplot(x='y', y='age', data=df)
plt.title("Distribución de Edad por Suscripción")
plt.show()
```
